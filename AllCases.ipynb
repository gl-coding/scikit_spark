```python
#!/usr/bin/python

import numpy as np
import scipy as sp
import urllib
import matplotlib.pyplot as pl

pl.figure(figsize=(20, 20))

cols=39
begin,end = 16,24
#begin,end = 1,30
categorycol=38

def cell(name, index):
    width = 3
    raw_data = urllib.urlopen(name)
    dataset = np.loadtxt(raw_data, delimiter=",", usecols=range(cols))
    x = dataset[:,0:end]
    y = dataset[:,categorycol]
    
    plt.subplot(width, width, index)
    pl.scatter(x[:,0], x[:,2],c=y)
    pl.autoscale(tight=True)
    pl.title(name)
    pl.xlabel("core")
    pl.ylabel("time")
    pl.grid()
    
def readListFromFile(file):
    fd = open(file)
    index = 0
    for line in fd.readlines():
        index += 1
        print line
        cell(line, index)
    
#cell("train", 1)
#cell("train", 2)
readListFromFile("filelist")
```
