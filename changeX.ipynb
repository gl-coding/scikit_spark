```python
#!/usr/bin/python
import os
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import urllib

plt.figure(figsize=(30, 30))

def error(f, x, y):
    return sp.sum((f(x)-y)**2)

def huatu(index, datax, datay, degree, labelx, labely):
    width = 6
    plt.subplot(width, width, index)
    plt.scatter(datax, datay)
    plt.autoscale(tight=True)
    plt.grid()
    
    fp1= sp.polyfit(datax, datay, degree)
    #print fp1
    #aa = polyval(fp1, 5)
    #print aa
    fStraight = sp.poly1d(fp1)
    #print "Error of Curve3 line:",error(fStraight,datax,datay)
    err = int(error(fStraight,datax,datay))
    #draw fitting straight line
    fx = sp.linspace(0,datax[-1], 10) # generate X-values for plotting
    plt.plot(fx, fStraight(fx), linewidth=4)
    plt.legend(["d=" + str(fStraight.order) + " err=" + str(err)], loc="lower right")
    plt.xlabel(labelx)
    plt.ylabel(labely)
    
def getArr(index):
    command = "awk -F\',\' \'{print $" + str(index) + "}\' title"
    res = os.popen(command).read().strip('\n')
    return res

print getArr(2)

cols=39
begin,end = 16,39
#begin,end = 1,30
categorycol=38
raw_data = urllib.urlopen("train.insert")
dataset = np.loadtxt(raw_data, delimiter=",", usecols=range(cols))
x = dataset[:,0:end]
y = dataset[:,categorycol]

aindex = 3
a = x[:,aindex]

bstart = 3

for i in range(1, 37):
    index = i
    bindex = bstart + i
    b = x[:,bindex]
    if bindex > 39:
        break
    huatu(index, b, a, 1, getArr(bindex), getArr(aindex))
```
